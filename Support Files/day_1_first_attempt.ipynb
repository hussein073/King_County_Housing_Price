```python
import statsmodels.api as sm
```

```python
df_bad[selected].head()
```

```python
df_bad_predictors = df_bad.drop(['price'], axis=1)
df_bad_predictors.head()
```

```python
df_bad_tgt = df_bad['price']
df_bad_tgt.head()
```

```python
# Set up predictors
predictors = sm.add_constant(df_bad_predictors)
```

```python
predictors.info()
```

```python
# Model SM LR model
model_sm = sm.OLS(df_bad_tgt, predictors).fit()

# View Results
model_sm.summary()
```

We can't have `NaN` values for the predictor variables! So for now, let's just remove the column(s) that have them and clean them further after this initial model.

```python
df_bad_predictors = df_bad.drop(['price','yr_renovated','waterfront'], axis=1)
df_bad_predictors.head()
```

```python
df_bad_predictors.info()
```

We've `drop`ped the columns with `NaN`s for now for our initial model (`waterfront` and `yr_renovated`).

```python
# Set up predictors
predictors = sm.add_constant(df_bad_predictors)

# Model SM LR model
model_sm = sm.OLS(df_bad_tgt, predictors).fit()

# View Results
model_sm.summary()
```

#### Initial observations
- $R^2$ (and $R^2_{adj}$) are not too bad at .564. But definitely could be improved on!
- All the variables are statistically significant ( p-values <0.05) with the exception of `zipcode`, which as a p-value of 0.054. Might look into dropping it.

```python
fig2 = plt.figure(figsize=(12,12))
fig = sm.graphics.plot_regress_exog(model_sm,"yr_built",fig=fig2)
plt.show()
```

```python
residuals = model_sm.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

#### Observations
qqplot shows that residuals are definitely not normal, so there are remaining issues with one or more of the following:  
- Outliers within `price` or other variables.
- Normality of `price` or other variables.
